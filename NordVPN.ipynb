```python
import requests
import os
import certifi
from bs4 import BeautifulSoup

# NordVPN configuration files URL
url = 'https://nordvpn.com/es/ovpn/'

# Define a function to download and save configuration files
def download_configs():
    # Set the SSL certificate authority (CA) bundle path
    cafile = certifi.where()
    # Send a GET request to the NordVPN configuration files URL
    response = requests.get(url, verify=cafile)

    # If the request was successful, parse the HTML content
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        # Find all links to .ovpn configuration files
        links = soup.find_all('a', href=lambda href: href.endswith('.ovpn'))
        # Create a directory to save the configuration files
        if not os.path.exists('nordvpn_configs'):
            os.mkdir('nordvpn_configs')
        # Download and save each configuration file
        for link in links:
            filename = os.path.basename(link['href'])
            filepath = os.path.join('nordvpn_configs', filename)
            with open(filepath, 'wb') as f:
                f.write(requests.get(link['href'], verify=cafile).content)
            print('Configuration file {} downloaded and saved to nordvpn_configs/{}'.format(filename, filename))
    else:
        print('Failed to download configuration files')

# Call the function to download and save the configuration files
download_configs()
```
